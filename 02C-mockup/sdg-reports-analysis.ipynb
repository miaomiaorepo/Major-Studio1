## Step1: scrape yearly reports
* libraries: requests, BeautifulSoup, pandas, times
* scrape each goal's pages separately
* 2018 is in different format

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
# making urls 
def get_url(keyword,year):
    url = "https://unstats.un.org/sdgs/report/"+year+"/"+keyword
    return url
```

```python
#grab text from each paragrahs
def get_para(url):
    response = requests.get(url)
    doc = BeautifulSoup(response.text,'html.parser')
    ps = doc.find(class_="page-content").find_all("p")
    
    #from <p>s, get and combine text
    paras = ''
    for i in ps:
        para = i.text
        paras = paras+para
    return paras
```

```python
keywords = ['Goal-16','Goal-13','Goal-14','Goal-01','Goal-02','Goal-03',
            'Goal-04','Goal-05','Goal-08','Goal-09','Goal-10','Goal-06',
            'Goal-07','Goal-11','Goal-12','Goal-15','Goal-17'
           ]
```

```python
# iterating every page for each goal to get one year data
# change year mannually
data = []
year = "2022"

for i in keywords:
    url = get_url(i,year)
    paras = get_para(url)
    
    result = {}
    result['text'] = paras
    result['goal'] = i
    result['year'] = year

    data.append(result)
```

```python
#get year 2018,it is in a different format
def para_2018(url):
    response = requests.get(url)
    doc = BeautifulSoup(response.text,'html.parser')
    #in 2018, all <p>s are under div with class "row"
    ps = doc.find(class_="row").find_all("p")
    
    #from <p>s, get and combine text
    paras = ''
    for i in ps:
        para = i.text
        paras = paras+para
    return paras
```

```python
#processing 2018
keywords_2018 = ['Goal-06',
            'Goal-07','Goal-11','Goal-12','Goal-15','Goal-17']

data = []
year = "2018"

for i in keywords_2018:
    url = get_url(i,year)
    paras = para_2018(url)
    
    result = {}
    result['text'] = paras
    result['goal'] = i
    result['year'] = year

    data.append(result)

df2018_half1= pd.DataFrame(data)
df2018_half1
```

## Step2: caculate word frequency
* libraries: punctuation, Counter
* remove punctuations and reduntant words
* count frequency

```python
from string import punctuation
from collections import Counter
```

```python
#get a whole year text
def get_report(report_name):
    ts=''
    for t in report_name.text:
        ts = ts+t
    return ts
```

```python
text2016 = get_report(df2016)
text2017 = get_report(df2017)
text2018 = get_report(df2018)
text2019 = get_report(df2019)
text2020 = get_report(df2020)
text2021 = get_report(df2021)
text2022 = get_report(df2022)
```

```python
# no mean words list
nomean = ['a','an','of','and','the','or','by','to','is','in',
          'that','as','per','for','its','it','was','were','are',
         'at','with','this','from','all','have','on','be',
         'been','which','this','has','more','than','such','had','but','–',
         'about','less','cent','most','they','their']
```

```python
def count_freq(text):
    
    tokens=[]
    
    #strip punctuations and convert all to lower cases
    for word in text.split():
        tokens.append(word.strip(punctuation).lower())
        
    #remove punctuations and no meaning words
    for i in nomean:
        for word in tokens:
            if word == i:
                tokens.remove(word)
    
    #count word frequency
    types = Counter(tokens)
    freq_list = []

    for word_freq in types.most_common(100):
            word_result = {}
            word_con, freq_con = word_freq
            word_result['word'] = word_con
            word_result['frequency'] = freq_con
            freq_list.append(word_result)
            
    df_freq = pd.DataFrame(freq_list)
    return df_freq
```

```python
df_word_2016 = count_freq(text2016)
df_word_2016
```

```python
df_word_2017 = count_freq(text2017)
df_word_2017
```

```python
df_word_2018 = count_freq(text2018)
df_word_2018
```

```python
df_word_2019 = count_freq(text2019)
df_word_2019
```

```python
df_word_2020 = count_freq(text2020)
df_word_2020
```

```python
df_word_2021 = count_freq(text2021)
df_word_2021
```

```python
df_word_2022 = count_freq(text2022)
df_word_2022
```

```python
whole_freq = pd.concat([df_word_2016,df_word_2017,df_word_2018,df_word_2019,df_word_2020,df_word_2021,df_word_2022], axis=1)
```

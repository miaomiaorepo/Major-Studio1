```python
import requests
```

```python
df = pd.read_excel("cleaned-death.xlsx")
```

```python
def gaode(location):  
        parameters = {
            'key': 'f2ca5eb449c189dc0ef99e178b5f9f63',
            'address': location,          
        }  
        base = 'http://restapi.amap.com/v3/geocode/geo?parameters' 
        response = requests.get(base, parameters)  
        answer = response.json()
        try:
            return answer['geocodes'][0]['location']
        except:
            return 0
```

```python
df['location']= df['GeoAreaName'].apply(gaode)
```
